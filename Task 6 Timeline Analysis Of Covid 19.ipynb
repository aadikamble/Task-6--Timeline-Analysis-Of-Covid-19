# Step 4 - Timeseries Analysis Of Covid 19

```python
df.total_cases.plot(figsize=(20,10),linewidth=5,fontsize=20)
plt.xlabel("Month",fontsize=20)
```

```python
df.new_cases.plot(figsize=(20,10),linewidth=5,fontsize=20)
plt.xlabel("Month",fontsize=20)
```

```python
total_deaths=df[['total_deaths']]
total_deaths.rolling(12).mean().plot(figsize=(20,10),linewidth=5,fontsize=20)
plt.xlabel("Month",fontsize=20)
```
